# Data Driven Segmentation 

```python
df4=pd.read_csv("2-wheeler-EV-bikewale.csv")
```

```python
df4.sample(2)
```

```python
df4.info()
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
import nltk
nltk.download('vader_lexicon')
sentiments = SentimentIntensityAnalyzer()
sentiments_list = []
for review_text in df4['review']:
    if pd.isna(review_text):
        sentiments_list.append('neutral')
        continue
    sentiment_scores = sentiments.polarity_scores(review_text)
    positive_score = sentiment_scores['pos']
    negative_score = sentiment_scores['neg']

    
    if positive_score > negative_score:
        sentiment_label = 'positive'
    elif positive_score < negative_score:
        sentiment_label = 'negative'
    else:
        sentiment_label = 'neutral'

    sentiments_list.append(sentiment_label)

df4['sentiment'] = sentiments_list
```

```python
df4
```

```python
df4_new=df4[['Used it for', 'Owned for', 'rating', 'Visual Appeal', 'Reliability','Performance', 'Maintenance cost',  'Service Experience', 'Comfort', 'sentiment']]
```

```python
df4_new.sample(4)
```

```python
df4_new = df4_new.fillna(0.)
```

```python
df4_new.info()
```

```python
from sklearn.preprocessing import StandardScaler, LabelEncoder
categorical_columns = ['Used it for', 'Owned for', 'sentiment']
numerical_columns = ['rating', 'Visual Appeal', 'Reliability','Performance','Maintenance cost',  'Service Experience', 'Comfort']
label_encoders = {col: LabelEncoder() for col in categorical_columns}

for col, encoder in label_encoders.items():
    df4_new[col] = encoder.fit_transform(df4_new[col])
scaler = StandardScaler()
all_columns_to_scale = categorical_columns + numerical_columns
data_scaled = scaler.fit_transform(df4_new[all_columns_to_scale])
df4_scaled = pd.DataFrame(data_scaled, columns=all_columns_to_scale)
```

```python
df4_scaled
```

# PCA 

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(random_state=42)
pca.fit(df4_scaled)
```

```python
data_pca = pca.transform(df4_scaled)
```

```python
df_pca = pd.DataFrame(data_pca, columns = [f'PC{x +1}' for x in range(len(df4_new.columns))])
```

```python
pca_summary = pd.DataFrame({'Standard Deviation':df_pca.std(),
                           'Proportion of Variance': pca.explained_variance_ratio_,
                           'Cumultive Proportion': pca.explained_variance_ratio_.cumsum()})
pca_summary
```

```python
instances = pd.DataFrame(pca.components_, columns = df4_new.columns, index = df_pca.columns).T
instances
```

```python
indx = instances.index

sns.scatterplot(data=df_pca, x = 'PC1', y = 'PC2', color = 'grey')
for i in range(len(instances['PC1'])):
    plt.arrow(0, 0, instances['PC1'][i]*5 , instances['PC2'][i]*5 , head_width=0.05, head_length=0.1, fc='red', ec='black')
    plt.text((instances['PC1'][i]*5), instances['PC2'][i]*5, indx[i], fontsize=12, ha='center', color = 'red', va = 'center_baseline')
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.title('PCA of the EV 2-Wheeler')
plt.grid(True)
plt.savefig("PCA.png")
plt.show()
```

# Extracting Segments 

```python
from sklearn.cluster import KMeans
from kneed import KneeLocator
```

```python
wcss = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(df4_scaled)
    wcss.append(kmeans.inertia_)

# Finding the optimal number of clusters using the KneeLocator
kneedle = KneeLocator(range(1, 11), wcss, curve='convex', direction='decreasing')
optimal_clusters = kneedle.elbow

# Plotting the elbow plot
plt.figure(figsize=(8, 6))
plt.plot(range(1, 11), wcss, marker='o', linestyle='-')
plt.xlabel("Number of Clusters")
plt.ylabel('WCSS')
plt.title("Elbow Method for Optimal Number of Clusters")
plt.axvline(optimal_clusters, linestyle='--', color='r', label=f'Optimal Clusters: {optimal_clusters}')
plt.legend()
plt.savefig("Kmean.png")
plt.show()

print(f"The optimal number of clusters is: {optimal_clusters}")
```

```python
optimal_clusters = 3  # Adjust this based on your analysis (e.g., using the elbow method)
kmeans = KMeans(n_clusters=optimal_clusters, random_state=42)
kmeans.fit_predict(df4_scaled)
df4_new['cluster'] = kmeans.labels_
```

```python
df4_new
```

Optimal number of cluster is 3 so number of segments will be 3

# Segment Profiling 

```python
data_clust = pd.DataFrame(df4_scaled, columns = df4_new.columns)
data_clust['cluster'] = kmeans.labels_
```

```python
columns_segment = ['Used it for', 'Owned for', 'rating', 'Visual Appeal', 'Reliability', 
                   'Performance', 'Maintenance cost', 'Service Experience', 'Comfort', 'sentiment']

# Ensure 'cluster' column is present and correctly defined
assert 'cluster' in df4_new.columns, "Cluster column not found in dataframe."

# Group by 'cluster' and calculate mean for each column
data_pivot1 = df4_new[columns_segment + ['cluster']].groupby('cluster').mean().T

# Calculate the overall mean for each variable across clusters
data_pivot_mean1 = data_pivot1.mean(axis=1).reset_index()
data_pivot_mean1.columns = ['Variable', 'Value']

# Plotting
plt.figure(figsize=(12, 9))

for i in range(3):  # Adjust based on the number of clusters
    plt.subplot(2, 2, i+1)
    sns.barplot(x=data_pivot1.columns[i], y=data_pivot1.index, data=data_pivot1, palette='viridis')
    sns.scatterplot(x='Value', y='Variable', data=data_pivot_mean1, color='red', s=100)
    
    for index, row in data_pivot_mean1.iterrows():
        plt.hlines(y=row['Variable'], xmin=0, xmax=row['Value'], colors='red', alpha=0.2)
    
    plt.ylabel("")
    plt.xlabel("")
    cluster_count = df4_new['cluster'].value_counts()[i]
    cluster_percentage = cluster_count * 100 / len(df4_new)
    plt.title(f"Cluster {i}: {cluster_count} ({cluster_percentage:.0f}%)")

plt.suptitle("Segment profile plot for the three-segment solution for the EV 2-Wheeler data set", fontsize=15)
plt.savefig("seg.png")
plt.tight_layout()
plt.show()
```

```python
def mosaicplot(df, color_list, figsize=(5,5), barwidth = 1, title = "Simple Mosaic Plot", xlabel = 'cat'):
    
    fig, ax = plt.subplots(figsize =figsize)
    
    # Convert to ratio of each row sum and do a cumsum along row
    p_cnt_ratio_df = df.div(df.sum(1), axis =0)
    ratio_cum_df = p_cnt_ratio_df.cumsum(axis=1)
    
    # Get column order.
    col_order = ratio_cum_df.columns.tolist()
        
    x = [str(n) for n in ratio_cum_df.index.tolist()]
    
    # Do the actual plots
    for col, colour in zip(col_order[::-1], color_list):
        plt.bar(x, ratio_cum_df[col], color=colour, edgecolor='white', width=barwidth, label=col)
        
    # Add in legend
    plt.legend(loc='upper left', bbox_to_anchor=(1,1), ncol=1)
    

    # Configure title and axis
    plt.title(title)
    plt.xlabel("Segments")
    ax.spines["right"].set_visible(False)
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    plt.yticks([])
    plt.show()
```

# Selection of Target Segment



The strategic target segments for the electric vehicle market are identified as Segment 1 (41% of consumers) and Segment 0 and 2 (57% of consumers). Segment 1's diverse preferences and dissatisfaction points present an opportunity for improving customer satisfaction and loyalty by directly addressing their specific demands. Segment 0 and segment 2  values visual appeal, reliability, service experience, and comfort, offering a chance to customize electric vehicles to meet these expectations and emphasize value for money. The strategy involves addressing dissatisfaction points in Segment 1 and enhancing positive elements in Segment 0 and segment 2, aligning electric vehicles with the distinct expectations of each segment to ensure competitive advantage and sustained market growth.
